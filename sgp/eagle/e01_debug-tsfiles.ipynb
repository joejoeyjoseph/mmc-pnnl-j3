```python
import pandas as pd
import glob
import re
import numpy as np
# pd.set_option('display.max_columns', 500)

def remove_duplicate_errorrow(ts_in_dir, ts_out_dir, run_hours, sim_id, turn): 

    ideal_file_num = run_hours*60*60/time_step_sec
    
    if turn == 'first': 
        target_files = '*.d02.*'
    if turn == 'days':
        target_files = '*_new'
    
    incorrect_files = []

    # WORKS
    # file = ts_in_dir+file_name
    for file in glob.glob(ts_in_dir+target_files):

        new_file = file+'_'+str(sim_id)

        lines_seen = set() # holds lines already seen
        outfile = open(new_file, 'w')

        for line in open(file, 'r'):

            if line not in lines_seen: # not a duplicate

                outfile.write(line)
                lines_seen.add(line)

        outfile.close()

        with open(new_file, 'r') as f:

            lines = f.readlines()

            if len(lines)-1 != ideal_file_num:

                print('--- wrong line numbers ---')
                print(file)
                print('file has', str(len(lines)))
                print('should be', str(ideal_file_num))
                
                def recur_test(new_file): 

                    try:

                        df = pd.read_csv(new_file, skiprows=1, header=None, delim_whitespace=True)

                    except pd.errors.ParserError as e:

                        errorstring = repr(e)
                        matchre = re.compile('Expected (\d+) fields in line (\d+), saw (\d+)')
                        (expected, line, saw) = map(int, matchre.search(errorstring).groups())

                        print('removing erroneous row:', str(line))
                        !sed -i "$line"'d' "$new_file"
                        
                        return recur_test(new_file)
                    
                recur_test(new_file)

                df = pd.read_csv(new_file, skiprows=1, header=None, delim_whitespace=True)

                if len(df) == ideal_file_num:
                    
                    print('corrected')

                    print('removing duplicated rows, moving it to the output directory')
                    !mv "$new_file" "$ts_out_dir"

                else: 
                    
                    print('**********')
                    print('WRONG HERE')
                    incorrect_files.append(new_file)

            else:

                print('moving', new_file)
                !mv "$new_file" "$ts_out_dir"
                
    return incorrect_files
```
